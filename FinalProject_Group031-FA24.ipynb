We want to find the relationship between life expectancy and schooling year, especially in countries with life expectancy below 65 years. This increase in educated professionals could improve healthcare continuity and alleviate workforce shortages by providing more qualified health professionals to fill critical roles. Thus, examining the relationship between schooling and life expectancy may reveal if educational investment could improve health outcomes in places where resources are scarce.

The glaring differences in life expectancies among countries and nations has led us to observe the relationship between these educational opportunities and how they can impact a country’s life expectancy. ## Vaccine Analysis

```python
#Vaccine data analysis
total = (data['Hepatitis_B'] + data['Polio'] + data['Diphtheria'] + data['Measles'])/4
#find mean vaccination rate
vacc = data.get(['Hepatitis_B', 'Polio', 'Diphtheria', 'Measles', "Life_expectancy"]).assign(total_vacc = total)
#Add as column to dataframe
vacc.describe()
```

Measles vaccine rates were the lowest on average, with a mean value of 77%. The maximum vaccination rate was 99% for any country. The lowest individual vaccination rates varied from 8% to 16%, but average vaccine rate was no lower than 30%. 

```python
#Plot total vaccination data histogram
plt.figure(figsize=(10, 6))
sns.histplot(vacc['total_vacc'], bins=20, kde=True)
plt.title('Distribution of Average Vaccination Rate')
plt.xlabel('Average Vaccination Rate')
plt.ylabel('Frequency')
plt.show()
```

The average vaccination rate is very left skewed, with 50% of the data falling in between 88 and 99, and the remaining 50% falling in frequency from 87 to 30.

```python
#get long form of data
long = pd.melt(vacc, id_vars="Life_expectancy", var_name="Vaccine", value_name="value")
#plot distribution of individual vaccines
plt.figure(figsize=(10, 6))
sns.histplot(long, x="value", hue = "Vaccine", bins=30, kde=True)
plt.title('Individual Vaccine Rate Distributions')
plt.xlabel('Vaccination Rate')
plt.ylabel('Frequency')
plt.show()
```

The distribution for all individual vaccines is also left skewed. Diphtheria and Polio follow a very similar distribution. Hepatitis is a little more varied, with a larger frequency peak before 90%. Measles features a very high peak around 63-65%. 

```python
#checking for year, country, or region bias in measles data
measle_60 = data[(data["Measles"] > 60) & (data["Measles"] < 65)].get(["Country", "Region", "Year"])
print(measle_60["Region"].value_counts())
print(measle_60["Year"].value_counts())
#Year is evenly distributed, region is not
print(measle_60["Country"].value_counts())
```

The spike in low Measles vaccination comes mostly from specific African countries. Interestingly, there is not a similar trend among all vaccines, only measles. Due to this spike, linear regression is not appropriate for Measles. The rest of the vaccines need to be normalized due to being skewed.

```python
#Transform skewed vaccine data
vacc["adjusted_vacc"] = (stats.boxcox(vacc["total_vacc"]))[0]
#Box-cox used because skew was too extreme for log/squared transformation
plt.figure(figsize=(10, 6))
sns.histplot(vacc['adjusted_vacc'], bins=10, kde=True)
plt.title('Distribution of Average Vaccination Rate')
plt.xlabel('Average Vaccination Rate')
plt.ylabel('Frequency')
plt.show()
```

While not as skewed as before, the data is still not completely normally distributed, with two peaks.

```python
#Graph scatterplot of adjusted vaccine rate and life expectancy
plt.figure(figsize=(20, 6))
sns.scatterplot(vacc, x = "adjusted_vacc", y = "Life_expectancy")
plt.title('Transformed Vaccine Rate vs Life Expectancy')
plt.xlabel('Box-Cox Transformed Average Vaccination Rate')
plt.ylabel('Life Expectancy (years)')
plt.show()
```

There does appear to be some sort of relationship between life expectancy and vaccine rates, but the data is noisy and a clear linear trend isn't visible.

```python
#Create vaccine vs life expectancy linear regression model
outcome, predictors = patsy.dmatrices("Life_expectancy ~ adjusted_vacc", vacc)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
residuals = results.resid
sns.scatterplot(vacc, x = "adjusted_vacc", y = residuals)
plt.title('Vaccine Regression Residual Plot')
plt.xlabel('Box-Cox Transformed Average Vaccination Rate')
plt.ylabel('Residuals')
plt.show()
```

While the p-value for the linear regression is low, the residual plot very clearly shows a pattern. This means linear regression is not appropriate for this data, and we cannot reject our null hypothesis that a country's vaccine rate does not have a positive effect on life expectancy.

```python
#setting up df for multivariate analysis
schoolvacc = vacc.get(["total_vacc","adjusted_vacc", "Life_expectancy"]).assign(schooling = data['Schooling'])
schoolvacc["logit_vacc"] = np.log(schoolvacc["total_vacc"]/((100-schoolvacc["total_vacc"])/100)) #logit function, to try and get it on the scale of schooling
schoolvacc.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(schoolvacc['logit_vacc'], bins=10, kde=True)
plt.title('Distribution of Average Vaccination Rate')
plt.xlabel('Average Vaccination Rate')
plt.ylabel('Frequency')
plt.show()
```

Next, we try using the logit function to normalize the vaccination rate data.

```python
#Graph scatterplot of (new) adjusted vaccine rate and life expectancy
plt.figure(figsize=(20, 6))
sns.scatterplot(schoolvacc, x = "logit_vacc", y = "Life_expectancy")
plt.title('Transformed Vaccine Rate vs Life Expectancy')
plt.xlabel('Logit Transformed Average Vaccination Rate')
plt.ylabel('Life Expectancy (years)')
plt.show()
```

```python
#Create logit vaccine rate vs life expectancy linear regression model
outcome, predictors = patsy.dmatrices("Life_expectancy ~ logit_vacc", schoolvacc)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
residuals = results.resid
sns.scatterplot(schoolvacc, x = "logit_vacc", y = residuals)
plt.title('Vaccine Regression Residual Plot')
plt.xlabel('Logit Transformed Average Vaccination Rate')
plt.ylabel('Residuals')
plt.show()
```

The residuals still seem to have a pattern, so we again cannot conclude anything about our hypothesis concretely. Perhaps schooling will give us a better idea of where to go. The schooling distribution is provided again for reference.

```python
#looking at schooling again
plt.figure(figsize=(10, 6))
sns.histplot(data['Schooling'], bins=10, kde=True)
plt.title('Distribution of Schooling')
plt.xlabel('Schooling')
plt.ylabel('Frequency')
plt.show()
```

```python
#Create schooling vs life expectancy linear regression model
outcome, predictors = patsy.dmatrices("Life_expectancy ~ schooling", schoolvacc)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
residuals = results.resid
sns.scatterplot(schoolvacc, x = "schooling", y = residuals)
plt.title('Schooling Regression Residual Plot')
plt.xlabel('Schooling in Years')
plt.ylabel('Residuals')
plt.show()
```

Performing OLS regression on schooling vs life expectancy gives a low p-value and randomized residual plot! However, is it possible to model life expectancy better with vaccine rate, for which we determined there was a possible positive correlation with life expectancy?

```python
#Create logit vaccine + schooling vs life expectancy linear regression model
outcome, predictors = patsy.dmatrices("Life_expectancy ~ logit_vacc + schooling", schoolvacc)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
residuals = results.resid
plt.figure(figsize=(12,6))
sns.scatterplot(schoolvacc, x = "logit_vacc", y = residuals, label = 'Adjusted Vaccine Rate', color = 'blue', alpha = 0.5)
sns.scatterplot(schoolvacc, x = "schooling", y = residuals, label = 'Schooling', color = 'red', alpha=0.5)
plt.title("Residuals vs Vaccine Rate and Schooling")
plt.xlabel("Logit Vaccine Rate (Blue) and Schooling (Red)")
plt.ylabel("Residuals")
plt.legend(title="Predictors", loc="upper left")
plt.show()
```

Using logit-adjusted average vaccination rate and schooling time in years yields an even better result: the residuals look even more randomized and the p-value is still low enough for us to reject the null hypothesis even with a significance level of α = 0.01. We can then conclude that there is a correlational relationship between vaccination rate+schooling time and life expectancy.

## Further Exploration of Data

Instead of just vaccination rate and schooling time, we wanted to see if a low infant mortality rate was possibly affecting our data. First, we checked the distribution of the infant mortality data.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Infant_deaths'], bins=10, kde=True)
plt.title('Distribution of Infant Mortality')
plt.xlabel('Deaths per 1000')
plt.ylabel('Frequency')
plt.show()
```

The data is very right skewed. To adjust, we used an exponential function.

```python
schoolvacc['infant'] = data["Infant_deaths"]
schoolvacc['infant_adjusted'] = (schoolvacc['infant'])**(0.5)
schoolvacc.describe()
```

Now that the data is much less skewed, we added it to our analysis. We looked at the Pearson correlation of our variables to see if our hypothesis is supported, or if there might be some other reason for our significant linear regression data.

```python
schoolvacc.select_dtypes(include='number').corr(method="pearson")
```

Earlier, we looked at two potential factors in life expectancy and found that schooling had an even larger effect than vaccines rates on life expectancy.<br>
logit_vacc: Coefficient = 2.1599, t-statistic = 18.119, p-value = 0.000 <br>
schooling: Coefficient = 1.7485, t-statistic = 40.929, p-value = 0.000<br>

However, after looking at the Pearson correlation between our variables, there appears to be a strong correlation between all three variables. While it's true that more years of schooling is associated with a higher life expectancy, the multicollinearity between them makes our analysis unreliable. Additionally, higher infant mortality rates are very correlated with lower life expectancy data. This makes sense, as a large amount of young deaths will lower the mean life expectancy of a country severely. 
To further improve our analysis, we would need to account for infant mortality skewing overall life expectancy, such as looking at median age lived or only looking at mortality among people older than 5. The correlation between variables in our dataset suggests that there might be an undiscovered underlying variable causing differences in both vaccine and schooling rates.

```python
#Create logit vaccine + schooling vs life expectancy linear regression model
outcome, predictors = patsy.dmatrices("Life_expectancy ~ logit_vacc + schooling + infant_adjusted", schoolvacc)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
residuals = results.resid
plt.figure(figsize=(12,6))
sns.scatterplot(schoolvacc, x = "logit_vacc", y = residuals, label = 'Adjusted Vaccine Rate', color = 'blue', alpha = 0.75)
sns.scatterplot(schoolvacc, x = "schooling", y = residuals, label = 'Schooling', color = 'red', alpha=0.75)
sns.scatterplot(schoolvacc, x = "infant_adjusted", y = residuals, label = 'Adjusted Infant Mortality Rate', color = 'green', alpha=0.75)
plt.title("Residuals vs Vaccine Rate and Schooling and Infant Mortality Rate")
plt.xlabel("Logit Vaccine Rate (Blue) and Schooling (Red) and Log Infant Mortality Rate (Green)")
plt.ylabel("Residuals")
plt.legend(title="Predictors", loc="upper left")
plt.show()
```

Performing Multiple Linear Regression on all three variables against life expectancy yields an r-squared value of 0.868, which shows a strong correlation between vaccination rate, schooling, and infant mortality rate.
